```python
def compute_cost(X, y, w, b):
    m = X.shape[0]
    cost = 0.0

    for i in range(m):
        y_pred = np.dot(X[i], w) + b
        err_i = (y_pred - y[i]) ** 2
        cost += err_i
    cost = cost/2 * m
    return cost
```

```python
def compute_gradient(X, y, w, b):
    m, n = X.shape
    dj_dw = np.zeros((n,))
    dj_db = 0

    for i in range(m):
        y_pred = np.dot(X[i], w) + b
        err_i = y_pred - y[i]
        for j in range(n):
            dj_dw[j] = dj_dw[j] + (err_i * X[i, j])
        dj_db += err_i
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    return dj_dw, dj_db
```

```python
def gradient_descent(X, y, w_ini, b_ini, alpha, iters):
    j_history = []
    w = w_ini
    b = b_ini

    for i in range(iters):
        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
        if i < 100000:
            j_history.append(compute_cost(X, y, w, b))

        if i% math.ceil(iters / 10) == 0:
            print(f"Iteration {i:4d} Cost {j_history[-1]:8.2f}")
            
    return w, b, j_history
```

```python
def polynomial_features(X, degree):
    X_poly = X.copy()
    for d in range(2, degree + 1):
        X_poly = np.hstack([X_poly, X**d])
    return X_poly
```

```python
X_poly = polynomial_features(X_normal, 2)
```

```python
m, n = X.shape
w_init = np.zeros(X_poly.shape[1])
b_init = 0
epoch = 100000
alpha = 1e-3
w_final, b_final, J_history = gradient_descent(X_poly, y_1, w_init, b_init, alpha, epoch)
print(f" Gradient descent found best w, b  at w: {w_final} and b:{b_final} ")

for i in range(m):
    print(f"Prediction: {(np.dot(X_poly[i], w_final) + b_final):.2f} Actual output: {y_1[i]}")
```
